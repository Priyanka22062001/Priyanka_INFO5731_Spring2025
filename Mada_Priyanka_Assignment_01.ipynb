# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math

print("r\tV")  # Table Heading as radius and Volume of Hemisphere corresponding to the radius
for r in range(2, 21, 2):  # Iterating through even numbers from 2 to 20
    V = (2/3) * math.pi * (r ** 3)  # Calculating volume of the hemisphere
    print(f"{r}\t{V:.2f}")  # Print radius and volume formatted to 2 decimal places
```

# Question 3 (10 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string

# Function to remove punctuation from the sentence
def remove_punctuation(sentence):
    """Remove punctuation from the sentence."""
    # Using str.translate() with str.maketrans to remove punctuation characters
    return sentence.translate(str.maketrans('', '', string.punctuation))

# Function to count the occurrences of each vowel in the sentence
def count_vowels(sentence):
    """Count the occurrences of each vowel in the sentence."""
    vowels = 'aeiou'  # Define the vowels
    # Create a dictionary that counts each vowel occurrence in the sentence
    vowel_count = {v: sentence.lower().count(v) for v in vowels}
    return vowel_count

# Main function to execute the program tasks
def main():
    # Task (a) - Prompt the user for two sentences
    sentence1 = input("Enter the first sentence: ")  # Prompt for the first sentence
    sentence2 = input("Enter the second sentence: ")  # Prompt for the second sentence

    # Task (b) - Convert sentences to lists of words
    words1 = sentence1.split()  # Split the first sentence into words
    words2 = sentence2.split()  # Split the second sentence into words
    # Print the combined list of words from both sentences
    print("Words from the two sentences:", words1 + words2)

    # Task (c) - Compare the two sentences and find common words
    common_words = list(set(words1).intersection(set(words2)))  # Find intersection of both word lists
    print("Common words in both sentences:", common_words)  # Print common words

    # Task (d) - Remove punctuation from sentences
    sanitized_sentence1 = remove_punctuation(sentence1)  # Remove punctuation from the first sentence
    sanitized_sentence2 = remove_punctuation(sentence2)  # Remove punctuation from the second sentence
    # Print the sanitized sentences with no punctuation
    print("Sanitized Sentence 1:", sanitized_sentence1)
    print("Sanitized Sentence 2:", sanitized_sentence2)

    # Task (e) - Count vowels in each sentence
    vowels_count1 = count_vowels(sanitized_sentence1)  # Count vowels in the first sentence
    vowels_count2 = count_vowels(sanitized_sentence2)  # Count vowels in the second sentence
    # Print the count of each vowel in both sentences
    print("Vowel counts in Sentence 1:", vowels_count1)
    print("Vowel counts in Sentence 2:", vowels_count2)

# Entry point of the program
if __name__ == "__main__":
    main()  # Call the main function to execute the program


```

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

```python
import math

# Function to evaluate the mathematical expression
def evaluate_expression(expression):
    try:
        # Evaluate the expression using Python's eval function
        # This supports addition, subtraction, multiplication, division, exponentiation, etc.
        result = eval(expression)
        return result
    except ZeroDivisionError:
        # Handle division by zero
        return "Error: Division by zero."
    except Exception as e:
        # Handle any other errors (e.g., invalid input)
        return f"Error: {e}"

# Main function to interact with the user and handle commands
def calculator():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, ** (exponentiation), sqrt (square root), % (modulus)")
    print("Type 'exit' to quit the program.")

    while True:
        # Prompt the user for input
        expression = input("Enter a mathematical expression: ")

        # If the user types 'exit', break the loop and quit the program
        if expression.lower() == 'exit':
            print("Exiting the calculator.")
            break

        # Extend the input to handle square root (sqrt) and ensure correct evaluation
        expression = expression.replace("sqrt", "math.sqrt")

        # Evaluate the expression and print the result
        result = evaluate_expression(expression)
        print(f"Result: {result}")

# Run the calculator
if __name__ == "__main__":
    calculator()  # Call the main function to execute the program
```

```python
import pandas as pd
import statistics

# Load the dataset
data_file = "Assignment1_denton_housing (1).csv"  # Ensure this path is correct
df = pd.read_csv(data_file)

# Rename columns for clarity
df.rename(columns={'year': 'Year', 'title_field': 'Title', 'value': 'Value'}, inplace=True)

# Convert 'Year' to integer and sort in descending order
df['Year'] = df['Year'].astype(int)
df.sort_values(by='Year', ascending=False, inplace=True)

# (a) Calculate year-to-year difference in Occupied Housing Units
diff_occupied = {}
occupied_df = df[df['Title'] == 'Occupied Housing Units'].sort_values(by='Year', ascending=False)
for i in range(len(occupied_df) - 1):
    year1, year2 = occupied_df.iloc[i]['Year'], occupied_df.iloc[i + 1]['Year']
    diff = occupied_df.iloc[i]['Value'] - occupied_df.iloc[i + 1]['Value']
    diff_occupied[f"{year2}-{year1}"] = diff

# Print differences in ascending order
print("Yearly Difference in Occupied Housing Units:")
for key, value in sorted(diff_occupied.items(), key=lambda x: x[1]):
    print(f"{key}: {value}")

# (b) Calculate percentage of vacant and occupied housing units
years = sorted(df['Year'].unique(), reverse=True)
print("\nYear   Vacant Housing Units    Occupied Housing Units")
for year in years:
    total_units = df[(df['Year'] == year) & (df['Title'] == 'Total Housing Units')]['Value'].values[0]
    vacant_units = df[(df['Year'] == year) & (df['Title'] == 'Vacant Housing Units')]['Value'].values[0]
    occupied_units = df[(df['Year'] == year) & (df['Title'] == 'Occupied Housing Units')]['Value'].values[0]
    vacant_percentage = round((vacant_units / total_units) * 100, 2)
    occupied_percentage = round((occupied_units / total_units) * 100, 2)
    print(f"{year} {vacant_percentage}% {occupied_percentage}%")

# (c) Find years with highest vacant and occupied housing units
max_vacant = df[df['Title'] == 'Vacant Housing Units'].nlargest(1, 'Value')
max_occupied = df[df['Title'] == 'Occupied Housing Units'].nlargest(1, 'Value')
print("\n                          Year    Value")
print(f"Vacant Housing Units      {max_vacant.iloc[0]['Year']}   {max_vacant.iloc[0]['Value']}")
print(f"Occupied Housing Units    {max_occupied.iloc[0]['Year']}   {max_occupied.iloc[0]['Value']}")

# (d) Calculate harmonic mean of total housing units
total_units_list = df[df['Title'] == 'Total Housing Units']['Value'].tolist()
harmonic_mean = statistics.harmonic_mean(total_units_list)
print(f"\nHarmonic Mean of Total Housing Units: {round(harmonic_mean, 2)}")
```

Develop a Product class that includes attributes such as product ID, name, category, quantity, and price. Additionally, create an Inventory class responsible for managing a collection of products. This class should contain methods that facilitate the addition, deletion, and updating of product details.

```python
import csv
import os

class Product:
    def __init__(self, product_id, name, category, quantity, price):
        self.product_id = product_id
        self.name = name
        self.category = category
        self.quantity = quantity
        self.price = price

    def __repr__(self):
        return f"Product({self.product_id}, {self.name}, {self.category}, {self.quantity}, {self.price})"


class Inventory:
    def __init__(self, filename='inventory.csv'):
        self.filename = filename
        self.products = []
        self.load_inventory()

    def load_inventory(self):
        if os.path.exists(self.filename):
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.reader(file)
                next(reader)  # Skip the header row
                for row in reader:
                    if row:  # Ensure the row is not empty
                        product_id, name, category, quantity, price = row
                        self.products.append(Product(product_id, name, category, int(quantity), float(price)))

    def save_inventory(self):
        with open(self.filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            # Write header
            writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price"])
            for product in self.products:
                writer.writerow([product.product_id, product.name, product.category, product.quantity, product.price])

    def add_product(self, product_id, name, category, quantity, price):
        if self.get_product(product_id):
            print(f"Product ID {product_id} already exists.")
            return
        new_product = Product(product_id, name, category, quantity, price)
        self.products.append(new_product)
        self.save_inventory()
        print(f"Product {name} added successfully.")

    def delete_product(self, product_id):
        product = self.get_product(product_id)
        if not product:
            print(f"Product ID {product_id} does not exist.")
            return
        self.products.remove(product)
        self.save_inventory()
        print(f"Product ID {product_id} deleted successfully.")

    def update_product(self, product_id, name=None, category=None, quantity=None, price=None):
        product = self.get_product(product_id)
        if not product:
            print(f"Product ID {product_id} does not exist.")
            return
        if name is not None:
            product.name = name
        if category is not None:
            product.category = category
        if quantity is not None:
            product.quantity = quantity
        if price is not None:
            product.price = price
        self.save_inventory()
        print(f"Product ID {product_id} updated successfully.")

    def get_product(self, product_id):
        for product in self.products:
            if product.product_id == product_id:
                return product
        return None

    def search_product(self, search_term):
        results = [product for product in self.products if search_term.lower() in product.name.lower() or
                   search_term.lower() in product.category.lower()]
        return results


# Example usage
if __name__ == "__main__":
    inventory = Inventory()

    # Adding products
    inventory.add_product("001", "Laptop", "Electronics", 10, 999.99)
    inventory.add_product("002", "Smartphone", "Electronics", 20, 499.99)

    # Searching for a product
    search_results = inventory.search_product("Laptop")
    print("Search Results:", search_results)

    # Updating a product
    inventory.update_product("001", quantity=8)

    # Deleting a product
    inventory.delete_product("002")
```

**PROMT 3:**
Design a method called restock_products to automatically restock items when their quantity falls below a predefined threshold. Additionally, implement a generate_report method to summarize item values, total quantities per category, and overall inventory value, reflecting any updates made.



```python
import csv
import os

class Product:
    def __init__(self, product_id, name, category, quantity, price):
        self.product_id = product_id
        self.name = name
        self.category = category
        self.quantity = quantity
        self.price = price

    def __repr__(self):
        return f"Product({self.product_id}, {self.name}, {self.category}, {self.quantity}, {self.price})"


class Inventory:
    def __init__(self, filename='inventory.csv', restock_threshold=5, restock_amount=10):
        self.filename = filename
        self.products = []
        self.restock_threshold = restock_threshold
        self.restock_amount = restock_amount
        self.load_inventory()

    def load_inventory(self):
        """Load inventory from a CSV file."""
        if os.path.exists(self.filename):
            try:
                with open(self.filename, mode='r', newline='') as file:
                    reader = csv.reader(file)
                    next(reader)  # Skip the header row
                    for row in reader:
                        if row:  # Ensure the row is not empty
                            product_id, name, category, quantity, price = row
                            self.products.append(Product(product_id, name, category, int(quantity), float(price)))
            except ValueError as e:
                print(f"Error reading data: {e}. The file may be corrupted.")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")
        else:
            print(f"File '{self.filename}' not found. Starting with an empty inventory.")

    def save_inventory(self):
        """Save current inventory to a CSV file."""
        try:
            with open(self.filename, mode='w', newline='') as file:
                writer = csv.writer(file)
                # Write header
                writer.writerow(["Product ID", "Name", "Category", "Quantity", "Price"])
                for product in self.products:
                    writer.writerow([product.product_id, product.name, product.category, product.quantity, product.price])
            print(f"Inventory saved to '{self.filename}' successfully.")
        except Exception as e:
            print(f"Error saving inventory: {e}")

    def add_product(self, product_id, name, category, quantity, price):
        """Add a new product to the inventory."""
        if self.get_product(product_id):
            print(f"Product ID {product_id} already exists.")
            return
        new_product = Product(product_id, name, category, quantity, price)
        self.products.append(new_product)
        self.save_inventory()
        print(f"Product {name} added successfully.")

    def delete_product(self, product_id):
        """Delete a product from the inventory."""
        product = self.get_product(product_id)
        if not product:
            print(f"Product ID {product_id} does not exist.")
            return
        self.products.remove(product)
        self.save_inventory()
        print(f"Product ID {product_id} deleted successfully.")

    def update_product(self, product_id, name=None, category=None, quantity=None, price=None):
        """Update details of an existing product."""
        product = self.get_product(product_id)
        if not product:
            print(f"Product ID {product_id} does not exist.")
            return
        if name is not None:
            product.name = name
        if category is not None:
            product.category = category
        if quantity is not None:
            product.quantity = quantity
        if price is not None:
            product.price = price
        self.save_inventory()
        print(f"Product ID {product_id} updated successfully.")

    def get_product(self, product_id):
        """Retrieve a product by its ID."""
        for product in self.products:
            if product.product_id == product_id:
                return product
        return None

    def search_product(self, search_term):
        """Search for products by name or category."""
        results = [product for product in self.products if search_term.lower() in product.name.lower() or
                   search_term.lower() in product.category.lower()]
        return results

    def restock_products(self):
        """Automatically restock items below the predefined threshold."""
        for product in self.products:
            if product.quantity < self.restock_threshold:
                product.quantity += self.restock_amount
                print(f"Restocked {product.name}: new quantity is {product.quantity}.")
        self.save_inventory()

    def generate_report(self):
        """Generate a report summarizing inventory."""
        total_value = 0
        category_summary = {}

        for product in self.products:
            total_value += product.price * product.quantity
            if product.category not in category_summary:
                category_summary[product.category] = 0
            category_summary[product.category] += product.quantity

        print("\n--- Inventory Report ---")
        print(f"Total Inventory Value: ${total_value:.2f}")
        print("Total Quantity per Category:")
        for category, quantity in category_summary.items():
            print(f"{category}: {quantity} items")
        print(f"Total number of products: {len(self.products)}")

# Example usage
if __name__ == "__main__":
    inventory = Inventory()

    # Adding products
    inventory.add_product("001", "Laptop", "Electronics", 3, 999.99)
    inventory.add_product("002", "Smartphone", "Electronics", 8, 499.99)
    inventory.add_product("003", "Desk Chair", "Furniture", 5, 89.99)

    # Restocking products
    inventory.restock_products()

    # Generating report
    inventory.generate_report()

    # Searching for a product
    search_results = inventory.search_product("Laptop")
    print("Search Results:", search_results)

    # Updating a product
    inventory.update_product("001", quantity=2)

    # Deleting a product
    inventory.delete_product("002")
```

```python
class Flight:
    def __init__(self, flight_number, destination, departure_time, total_seats):
        self.flight_number = flight_number
        self.destination = destination
        self.departure_time = departure_time
        self.total_seats = total_seats
        self.reserved_seats = 0

    def __str__(self):
        return (f"Flight {self.flight_number}: {self.destination}, "
                f"Departure: {self.departure_time}, "
                f"Total Seats: {self.total_seats}, "
                f"Reserved Seats: {self.reserved_seats}")


class FlightBookingSystem:
    def __init__(self):
        self.flights = []

    def add_flight(self, flight_number, destination, departure_time, total_seats):
        if any(flight.flight_number == flight_number for flight in self.flights):
            print(f"Flight {flight_number} already exists.")
            return

        new_flight = Flight(flight_number, destination, departure_time, total_seats)
        self.flights.append(new_flight)
        print(f"Flight {flight_number} added successfully.")

    def find_flight(self, flight_number):
        for flight in self.flights:
            if flight.flight_number == flight_number:
                return flight
        print(f"Flight {flight_number} not found.")
        return None

    def reserve_seat(self, flight_number, num_seats):
        flight = self.find_flight(flight_number)
        if flight:
            if num_seats <= 0:
                print("Invalid number of seats to reserve. Must be greater than zero.")
                return

            if flight.reserved_seats + num_seats > flight.total_seats:
                print(f"Cannot reserve {num_seats} seats. Only {flight.total_seats - flight.reserved_seats} seats available.")
            else:
                flight.reserved_seats += num_seats
                print(f"{num_seats} seats reserved on flight {flight_number}. Total reserved: {flight.reserved_seats}")

    def release_seat(self, flight_number, num_seats):
        flight = self.find_flight(flight_number)
        if flight:
            if num_seats <= 0:
                print("Invalid number of seats to release. Must be greater than zero.")
                return

            if num_seats > flight.reserved_seats:
                print(f"Cannot release {num_seats} seats. Only {flight.reserved_seats} seats are currently reserved.")
            else:
                flight.reserved_seats -= num_seats
                print(f"{num_seats} seats released on flight {flight_number}. Total reserved: {flight.reserved_seats}")

    def cancel_reservation(self, flight_number, num_seats):
        flight = self.find_flight(flight_number)
        if flight:
            if num_seats <= 0:
                print("Invalid number of seats to cancel. Must be greater than zero.")
                return

            if num_seats > flight.reserved_seats:
                print(f"Cannot cancel {num_seats} seats. Only {flight.reserved_seats} seats are currently reserved.")
            else:
                flight.reserved_seats -= num_seats
                print(f"{num_seats} seats canceled on flight {flight_number}. Total reserved: {flight.reserved_seats}")

    def generate_report(self):
        total_flights = len(self.flights)
        total_available_seats = 0
        overbooked_flights = []

        for flight in self.flights:
            available_seats = flight.total_seats - flight.reserved_seats
            total_available_seats += available_seats

            if flight.reserved_seats > flight.total_seats:
                overbooked_flights.append(flight.flight_number)

        print("\n--- Flight Performance Report ---")
        print(f"Total Flights: {total_flights}")
        print(f"Total Available Seats: {total_available_seats}")
        if overbooked_flights:
            print(f"Overbooked Flights: {', '.join(overbooked_flights)}")
        else:
            print("No Overbooked Flights.")

# Example Usage
if __name__ == "__main__":
    booking_system = FlightBookingSystem()

    # Adding flights
    booking_system.add_flight("AA123", "New York", "2025-03-15 10:00", 150)
    booking_system.add_flight("BB456", "Los Angeles", "2025-03-16 12:00", 200)

    # Reserving seats
    booking_system.reserve_seat("AA123", 10)
    booking_system.reserve_seat("AA123", 145)  # Should fail due to overbooking

    # Canceling reservations
    booking_system.cancel_reservation("AA123", 5)
    booking_system.cancel_reservation("AA123", 10)  # Should fail since only 5 are reserved

    # Generating a performance report
    booking_system.generate_report()
```
